```python
from math import atan2, sqrt, degrees

u = (3, 4)
v = (4, 3)

def length(v):
    return sqrt(v[0]**2 + v[1]**2) 

def to_polar(vector):
    x, y = vector[0], vector[1]
    angle = atan2(y,x)
    return (length(vector), angle)

u_polar = to_polar(u)
v_polar = to_polar(v)

u_polar[1] - v_polar[1]



```
